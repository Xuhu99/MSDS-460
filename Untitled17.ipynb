```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Define edges: (from, to, weight)
edges = [
    ('Origin', 'A', 40), ('Origin', 'B', 60), ('Origin', 'C', 50),
    ('A', 'B', 10), ('A', 'D', 70),
    ('B', 'C', 20), ('B', 'E', 40),
    ('C', 'D', 50), ('C', 'E', 40),
    ('D', 'E', 10), ('D', 'Destination', 60),
    ('E', 'Destination', 80)
]

# Create a directed graph
G = nx.DiGraph()
G.add_weighted_edges_from(edges)
```

```python
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Town Map - Distance Graph")
plt.show()
```

```python
start = 'Origin'
end = 'Destination'

path = nx.dijkstra_path(G, start, end)
length = nx.dijkstra_path_length(G, start, end)

print("Shortest path:", "  ".join(path))
print("Total distance:", length, "miles")
```

```python
print(f"Total cost: ${length}")
print(f"Total time: {length} minutes")
```
